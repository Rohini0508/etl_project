```python
#scd2
scd2_df=file_path.copy()
```

```python
scd2_df['start_date']= pd.to_datetime('2024-01-01')
scd2_df['end_Date']=pd.NaT
scd2_df['is_current'] = True 

```

```python
from datetime import datetime, date, time
# Now simulate change: customer_id = 1 loyalty status changed
new_loyalty = 'Platinum'
change_date = datetime(2025, 6, 24)
# Step 1: End the current version
scd2_df.loc[
    (scd2_df['customer_id'] == 1) & (scd2_df['is_current']),
    ['end_date', 'is_current']
] = [change_date, False]

# Step 2: Insert a new row with updated info
new_row = scd2_df[scd2_df['customer_id'] == 1].iloc[0].copy()
new_row['loyalty_status'] = new_loyalty
new_row['start_date'] = change_date
new_row['end_date'] = pd.NaT
new_row['is_current'] = True

scd2_df = pd.concat([scd2_df, pd.DataFrame([new_row])], ignore_index=True)
```

```python
#Scd4
current_table = file_path.copy()
history_table = pd.DataFrame(columns=file_path.columns.tolist() + ['change_date'])

# Simulate a change for customer_id = 1
customer_id = 1
change_date = datetime(2025, 6, 24)

# Step 1: Archive current row to history
old_row = current_table[current_table['customer_id'] == customer_id].copy()
old_row['change_date'] = change_date

history_table = pd.concat([history_table, old_row], ignore_index=True)

# Step 2: Update current table
current_table.loc[current_table['customer_id'] == customer_id, 'loyalty_status'] = 'Platinum'
```
